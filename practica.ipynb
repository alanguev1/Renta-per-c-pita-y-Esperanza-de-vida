```python
# importamos numpy para calcular el logaritmo
import numpy as np

# crear una nueva variable con el logaritmo de otra
df["log_renta_pc"] = np.log(df["Renta per capita"])
plt.figure(figsize=(14.5, 10))

# cambiamos la variable independiente y les damos un parámetro para utilizarlas después
x=df["log_renta_pc"]
y=df["Esperanza de vida"]

plt.scatter(data=df, x="log_renta_pc", y="Esperanza de vida", s="Poblacion normalizada", c="Poblacion normalizada", 
            alpha=0.5)

plt.title("La esperanza de vida está relacionada con la renta per cápita", fontsize=15)

# modificamos la etiqueta
plt.xlabel("Renta per capita (log)", fontsize=15)
plt.ylabel("Esperanza de vida", fontsize= 15)
plt.figtext(0.6, 0.2, "Ajustamos a escala semilogarítmica", fontsize= 15)


df.reset_index(drop=True, inplace=True)
for i in range(0, 10):
    plt.annotate(df['País'][i], (x[i], y[i]), fontsize=10, weight="bold")

plt.savefig('6.png')
```

```python
plt.figure(figsize=(14.5, 10))
x=df["log_renta_pc"]
y=df["Esperanza de vida"]

plt.scatter(data=df, x="log_renta_pc", y="Esperanza de vida", s="Poblacion normalizada", c="Poblacion normalizada", 
            alpha=0.5)


plt.title("La esperanza de vida está relacionada con la renta per cápita", fontsize=15)
plt.xlabel("Renta per capita (log)", fontsize=15)
plt.ylabel("Esperanza de vida", fontsize= 15)
plt.figtext(0.6, 0.2, "Añadimos línea de tendencia", fontsize= 15)

df.reset_index(drop=True, inplace=True)
for i in range(0, 10):
    plt.annotate(df['País'][i], (x[i], y[i]), fontsize=10, weight="bold")

# definimos m como la pendiente y b como la intercepción mediante un fit polinomial, en este caso y=b+m*x
m, b = np.polyfit(x, y, 1)

#dibujamos la línea
plt.plot(x, m*x+b, color='black')

plt.savefig('7.png')
```

```python
# importamos statsmodel (se parece más a R y creo que es mejor para realizar regresiones)
import statsmodels.api as sm

# Definimos variables y añadimos la constante

Y = df["Esperanza de vida"]
X = sm.add_constant(df["log_renta_pc"])
W = df["Poblacion"]

# modelo de regresión lineal con residuos ponderados por población
model = sm.WLS(Y, X, weights = W).fit()
model.summary()
```

```python
#11.1. #Estimamos predicción y residuos al cuadrado de cada observación
df["Prediction"] = model.predict(X)

df["Residuals"] = abs(df["Residuals"])
       
df.tail()
```

```python
sm.qqplot(model.resid, line ='45')
plt.title("El modelo está muy lejos de la normalidad, pese a la transformación logarítimica");
```

```python
plt.scatter(model.predict(X), model.resid)
plt.axhline(np.mean(model.resid), color="red")
plt.title("Los residuos parecen distribuirse aleatoriamente")
plt.xlabel("predicciones")
plt.ylabel("residuos");
```

```python
df1 = df[["País", "Residuals"]].sort_values("Residuals", ascending=False).head(10)
df1.reset_index(drop=True, inplace=True)

plt.figure(figsize=(20, 10))
plt.bar(df1["País"], df1["Residuals"])
plt.title("Todos los países más alejados de la predicción son africanos", fontsize=20)
plt.xlabel("Países")
plt.ylabel("Residuos absolutos");
```
